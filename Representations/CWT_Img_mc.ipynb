#### Imagen de escalograma con CWT de EEG de mental calculation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pywt
import cv2
```

```python
# importar un canal de una muestra
# canal C4 es el 19
# 18*1344 = 24192, 19*1344 = 25536
# 18*1024 = 18432, 19*1024 = 19456
# 18*1344+160 = 24352, 19*1344-160 = 25376
canalC4 = pd.read_csv("MeCa_car_10_1.csv", header=None).iloc[0, 24192:25536]
# canalC4 = pd.read_csv("VentaneoMC-01-1.csv", header=None).iloc[1, 18432:19456]
```

```python
canalC4.shape
```

```python
# 2.625 segundos con fs=512Hz
fs = 512
time = np.arange(0, 2.625, 1/fs)
sig = canalC4
print(sig.shape)

plt.plot(time, sig)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# calcular las escalas de 8-30Hz con pasos de 0.5Hz para complex Morlet wavelets
# f = scale2frequency(wavelet, scale)/sampling_period
escalas = pywt.scale2frequency('cmor3-3', np.arange(8,30.5,0.5)) / (1/fs)
print(escalas)
```

```python
# Escalograma (CWT)
# pywt.cwt(senal, escalas, tipo wavelet, periodo de muestreo = (1 / fs))
coef, freqs = pywt.cwt(sig, escalas, 'cmor3-3', sampling_period = (1 / 512))
```

```python
# dado que cmor3-3 genera numeros complejos, calcular el modulo
# la formula del modulo se presenta en la siguiente celda
x = abs(coef)
```

$|z| = |a + bi| = \sqrt{a^2 + b^2}$

```python
plt.pcolormesh(time, freqs, x)
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
# esto no se utilizo en la tesis, es un trabajo futuro
# imprimir aplicando ln para resaltar el color
plt.pcolormesh(time, freqs, np.log(x))
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```

```python
print(x.shape)
plt.matshow(x)
plt.show()
```

```python
# resize proporcional de dos con interpolacion inter area
# reducir el tamaño entre 2
dim = (int(np.floor(x.shape[1]/2)), int(np.floor(x.shape[0]/2))) # ancho, alto
resized = cv2.resize(x, dim, interpolation=cv2.INTER_AREA)
print(resized.shape)
plt.imshow(resized)
plt.show()
# otra manera de imprimir la imagen
plt.matshow(resized)
plt.show()
```

```python
# ln de CWT, para resaltar los colores
x_log = np.log(resized)
```
